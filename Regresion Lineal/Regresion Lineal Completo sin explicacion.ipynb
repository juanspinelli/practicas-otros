```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

boston = load_boston()
numero_habitaciones = boston.data[:, 5]
valor_medio = boston.target

x = np.array(numero_habitaciones)
y = np.array(valor_medio)
```

Formula minimizar el error cuadratico medio ( Estimacion minimos cuadrados ordinario)

$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
plt.scatter(x, y, alpha = 0.3)
x = np.array([np.ones(506),x]).T
B = np.linalg.inv(x.T @ x) @ x.T @ y
### B[0] es el termino independiente
plt.plot([3, 10], [B[0] + B[1] * 3, B[0] + B[1] * 10], c="red")
```

### Ahi esta la linea de regresion lineal estimada por los minimos cuadrados ordinarios.
