## Definir tranpuesta de X
B = (x.T)
Forma convencional
B = (np.transpose(x))

B = (x.T @ x)

@ Significa multiplicacion matricial

Ahora hay que invertir, se usa numpy y del paquete de algebra lineal la funcion inv

Esto seria la inversa de la matriz x tranpuesta de x
B = np.linalg.inv(x.T @ x)

Ahora esto se realiza la multiplicacion matricial denuevo por la matriz X de T y luego por la matriz Y

B = np.linalg.inv(x.T @ x) @ X.T @ Y
