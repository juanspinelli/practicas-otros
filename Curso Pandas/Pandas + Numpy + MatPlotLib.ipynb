```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=['A', 'B', 'C', 'D'])
```

```python
df = df.cumsum()
```

```python
df.plot()
```
