```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data = pd.read_csv('archivos/Libro1.csv', header = 0, sep=';')
g = (data.groupby(["SUBDETERMINANTE"])["POSITIVAS", "POSIBLES"]
     .sum()
     .sort_values(by="POSITIVAS"))
g.plot(kind='barh', legend='Reverse')
plt.title(u'ANALISIS DE SUBDETERMINANTES', size = 12.0)
plt.xlabel(u'Suma', size = 14.0)
plt.ylabel('SubDeterminantes', size = 12.0)
fig_size = plt.rcParams["figure.figsize"]
print ("Current size:", fig_size)
fig_size[0] = 15.0
fig_size[1] = 15.0
plt.savefig('detalle', bbox_inches='tight')
```

```python
a = 2
m = 10

def subtask_service_time():
    return (np.random.pareto(a) + 1) * m

print('Simulation mean:', sum([subtask_service_time() for _ in range(1000)]) / 1000)
print('Excepted mean:', a * m / (a - 1))
```

```python
datosDos = {'Nombre': ['Juan', 'Lucas', 'Chato', 'Damian', 'Romina', 'N/A'], 
         'Calificaciones': ['99', '99', np.nan, '98', '98', '97'], 
         'Deportes': ['Futbol','Futbol','N/A','Futbol','Voley','Futbol'], 
         'Materias': ['Algebra', 'Javascript', 'Python', 'N/A', 'After Effect', 'Premiere']}
dfDos = pd.DataFrame(datosDos)
print(dfDos)
print('\n' * 2)
```

```python
print(dfDos.info())
print('\n' * 2)
```

```python
print(dfDos.describe())
print('\n' * 2)
```

```python
sinNaN = pd.DataFrame(dfDos)
sinNaN = sinNaN.replace(np.nan, '0')
print(sinNaN)
print('\n' * 2)
```

```python
print(sinNaN.describe())
print('\n' * 2)
```

```python
### Quitar toda la linea donde haya un NaN en lugar de replace
eliminarLineaNaN = pd.DataFrame(dfDos)
eliminarLineaNaN.dropna(how='any', inplace = True)
print(eliminarLineaNaN)
print('\n' * 2)
```

```python
### Eliminar registro buscando por columna
columna = dfDos[dfDos['Nombre'] != 'N/A']
print(columna)
print('\n' * 2)
```

```python
### Llenar con 0 cualquier valor con NaN
ponerCero = pd.DataFrame(dfDos)
ponerCero.fillna(0, inplace = True)
print(ponerCero)
print('\n' * 2)
```

```python
print(ponerCero.describe())
print('\n' * 2)
```

```python
cambiarStringInt = pd.DataFrame(dfDos)
cambiarStringInt['Calificaciones'] = cambiarStringInt.Calificaciones.astype(int)
print(cambiarStringInt.describe())
print('\n' * 2)
```

```python
### Estadisticas individuales
print('Promedio', cambiarStringInt['Calificaciones'].mean())
print('Minimo', cambiarStringInt['Calificaciones'].min())
print('Maximo', cambiarStringInt['Calificaciones'].max())
print('\n' * 2)
```
