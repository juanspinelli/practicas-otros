```python
import requests
import urllib.request
import pandas as pd
import re
import json
from bs4 import BeautifulSoup

url = 'https://horoscoponegro.com/horoscopo-semanal'
path_file_dest = r'horoscopo-semanal.json'
url_inicial = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
zodiac_signs_final = []

def read_site(site_url):
    site = requests.get(site_url, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
    if site.status_code == 200:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        return soup
    else:
        print('•  La página no responde')

def write_doc_json(content, mode):
    own_file = open(path_file_dest, mode, encoding='utf-8')
    own_file.write(content.decode('utf-8'))
    own_file.close()
    print('     El archivo json fue sobre-escrito y se creo el df data')

print('•  Verificando conectividad a la url solicitada ...')

if url_inicial.status_code == 200:
    print('     Acceso Ok ... \n')
    print('•  Verificando sub urls ...')
    horoscopos = pd.DataFrame()
    
    urls = []
    soup = BeautifulSoup(url_inicial.text, "html.parser")
    
    for link in soup.find_all('a', {'href': re.compile('-semanal-')}):
        urls.append(link.get('href'))
        
    for x in range(0, len(urls)):
        url_sub = urls[x]
        acceso = requests.get(url_sub, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
        if acceso.status_code == 200:
            print('     Url ' + str(x + 1) + ' de ' + str(len(urls)) + ' obtenida con exito!')
            soup = BeautifulSoup(acceso.text, "html.parser")
            title = soup.find('h1', {'class': 'entry-title'}).get_text()
            week = soup.find('h2').get_text()
            
            text = ''
            
            for element in soup.find_all(['p', 'strong']):
                text += element.get_text()
                
            text = text.replace(week, '')
            text = text.replace('[email\xa0protected]', '')
            
            individual = {'title' : title, 'week' : week, 'text' : text}
            zodiac_signs_final.append(individual)
        
    content = json.dumps(zodiac_signs_final, indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
    print('\n•  Creando/sobre-escribiendo archivo json')
    write_doc_json(content ,'w+')
    data = pd.read_json(path_file_dest, encoding = 'UTF-8')

else:
    print('•  Error al tratar de conectar')
```
