```python
import requests
from bs4 import *
import json
import pandas as pd
from IPython.display import Image, display, Markdown
import glob
```

```python
url = 'https://www.ranker.com/list/most-disturbing-thanos-comics/zack-howe'
path_file_dest = r'thanos.json'

def write_doc_json(content, mode):
    own_file = open(path_file_dest, mode, encoding='utf-8')
    own_file.write(content.decode('utf-8'))
    own_file.close()
    print('\n  El archivo json fue sobre-escrito con exito!')

urls = []
for x in range(1, 5):
    urls.append(url + '?page=' + str(x))
    
thanos = []

for x in urls:
    
    response = requests.get(x, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
    
    if response.status_code == 200:
        
        soup = BeautifulSoup(response.text, "html.parser")
        title = soup.find_all('span', {'class': 'listItem__h2Link'})
        img = soup.find_all('img', {'class' : 'listItem__image listItem__image--blog'})
        text = soup.find_all('meta', {'itemprop' : 'description'})
        

        for i in range (0, len(title)):
            
            info = {'title' : title[i].get_text(), 
                    'img' : img[i]['src'],
                    'text' : text[i]['content']
                   }
            
            thanos.append(info)
        
    else:
        
        print('Error al conectar a la direccion: ' + str(x))

content = json.dumps(thanos, indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
write_doc_json(content ,'w+')
data = pd.read_json(path_file_dest, encoding = 'UTF-8')

print('  Se leyo el json y se creo el dataframe data')

data.head()
```

```python
k = 1
for x in range (0, data.shape[0]):
    name_image_folder = 'thanosImg/' + str(k) + '.jpg'
    image = requests.get(data.img[x]).content
            
    with open(name_image_folder, 'wb') as handler:
            handler.write(image)
    k += 1
    
class bcolors:
    BOLD = '\033[1m'
    ENDC = '\033[0m'

imgs =len(glob.glob("thanosImg/*.jpg"))

listOfImageNames = []

for x in range(1, imgs + 1):
    listOfImageNames.append('thanosImg/' + str(x) + '.jpg')

k = 0

for imageName in listOfImageNames:
    print('\n' * 2)
    print(bcolors.BOLD + data.title[k] + bcolors.ENDC)
    print('')
    print(data.text[k])
    print('')
    display(Image(filename=imageName))
    k += 1
```
